```python
import pytoulbar2
```

### Case 1:  Unconstrained Square

<img src="ex_unc.png"  width="360" height="360">

```python
# problem definition
top = 16
Problem1 = pytoulbar2.CFN(top)

# variable definitions
for i in range(2):
    Problem1.AddVariable('X'+str(i+1), [0, 1])

# position costs
Problem1.AddFunction([0, 1], [5, 2, 4, 3])

# solve
res1 = Problem1.Solve(showSolutions=1)
if res1:
    print(res1)
```

### Case 2:  One Constraint

<img src="ex_c1.png"  width="360" height="360">

```python
# problem definition
top = 16
Problem2 = pytoulbar2.CFN(top)

# variable definitions
for i in range(2):
    Problem2.AddVariable('X'+str(i+1), [0, 1])

# position costs
Problem2.AddFunction([0, 1], [5, 2, 4, 3])

# constraints
Problem2.AddLinearConstraint([2, -2], [0, 1], '>=', -1)

# solve
res2 = Problem2.Solve(showSolutions=1)
if res2:
    print(res2)
```

### Case 3:  Two Constraints

<img src="ex_c2.png"  width="360" height="360">

```python
# problem definition
top = 16
Problem3 = pytoulbar2.CFN(top)

# variable definitions
for i in range(2):
    Problem3.AddVariable('X'+str(i+1), [0, 1])

# position costs
Problem3.AddFunction([0, 1], [5, 2, 4, 3])

# constraints
Problem3.AddLinearConstraint([2, -2], [0, 1], '>=', -1)
Problem3.AddLinearConstraint([1, 3], [0, 1], '<=', 2)

# solve
res3 = Problem3.Solve(showSolutions=1)
if res3:
    print(res3)
```
